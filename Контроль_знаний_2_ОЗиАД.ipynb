```python
 #Контроль знаний 2-3. Методы статистического анализа данных. Нормализация и стандартизация данных временных рядов
 #Вариант 3 Алжир 01.12.2010-01.12.2020
```

```python
from pandas import read_csv, read_html
from sklearn.preprocessing import MinMaxScaler
df = pd.read_html(
  'http://pogoda-service.ru/archive_gsod_res.php?country=AG&station=606770&datepicker_beg=01.12.2010&datepicker_end=01.12.2020&bsubmit=%D0%9F%D0%BE%D1%81%D0%BC%D0%BE%D1%82%D1%80%D0%B5%D1%82%D1%8C',encoding='utf-8',header=0,index_col=0)
series = df[0]
#type(series)
```

```python
print((series.head()))
```

```python
values = series.values
values = values.reshape((7*len(values), 1))
scaler = MinMaxScaler(feature_range=(0,1))
scaler = scaler.fit(values)
print('Min: %f, Max: %f' % (scaler.data_min_,scaler.data_max_))
```

```python
normalized = scaler.transform(values)
for i in range(5):
  print(normalized[i])
```

```python
inversed = scaler.inverse_transform(normalized)
for i in range(5):
  print(inversed[i])
```
