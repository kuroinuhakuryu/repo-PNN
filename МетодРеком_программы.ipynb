```python
#Решение
import random
import copy
import ast
#Вычисление возможных сумм
def get_sum_list(monet_list):
  result = set()
  if len(monet_list) == 1:
    return [monet_list[0]]
  new_list = copy.deepcopy(monet_list)
  m = new_list[0]
  new_list.remove(m)
  res_sum = get_sum_list(new_list)
  result.add(m)
  for r in res_sum:
    result.add(r+m)
    result.add(r)
  return result
#Сопоставление возможных сумм, с учетом сдачи
def add_minus_list(monet_list, first_list):
  result = set()
  new_list = copy.deepcopy(monet_list)
  m = new_list[0]
  new_list.remove(m)
  if len(new_list)>0:
    res_sum = add_minus_list(new_list, first_list)
  else:
    res_sum = list(first_list)
  for r in res_sum:
    if r-m>0:
      result.add(r-m)
    result.add(r)
  return result

#Вычисление минимальных сумм фактической покупки
def get_min_calc(S):
  new_list = copy.deepcopy(S)
  m = new_list[0]
  new_list.remove(m)
  if len(new_list) == 0:
    return [{0},{m}]
  res_sum = get_min_calc(new_list)
  res_sum.append(set())
  for k in reversed(range(len(res_sum)-1)):
#if res_sum[k+1]<res_sum[k]+m:
    for r in res_sum[k]:
      res_sum[k+1].add(r+m)
  return res_sum

#Получение минимальных сумм и максимизация покупок
def get_max_items(F,S):
  for k in reversed(range(len(S))):
# print(S[k])
    min_sum = max(F)+1
    for p in S[k]:
      if p in F and p<min_sum:
        min_sum = p
      if min_sum<=max(F):
        return k, min_sum
  return 0,0

#Обработка данных и решение
def solve(dataset):
  input_ds = dataset[1:-2].split("],[")
  if len(input_ds)<3:
    print(dataset)
  return dataset
  M = [int(i) for i in input_ds[0].split(",")]
  N = [int(i) for i in input_ds[1].split(",")]
  S = [int(i) for i in input_ds[2].split(",")]
  f = get_sum_list(M)
  r = add_minus_list(N,f)
  s = get_min_calc(S)
  k, mins = get_max_items(r,s)

  return str(mins)
```

```python
#Задача 5. Решение
import random
import math
import ast
#получить кластер от конкретного участника
def getcluster_from_point(r,p,cluster=[]):
  cluster = list(set(cluster+[p]))
  new_ps = [i for i in range(len(r[0])) if r[p][i]==1 and i not in cluster]
  cluster = list(set(cluster+new_ps))
  for new_p in new_ps:
    clustern = getcluster_from_point(r,new_p, cluster=cluster)
    cluster = list(set(clustern))
  return cluster
#получить все кластера
def get_clusters(r):
  points = [i for i in range(len(r[0]))]
  clusters = []
  while len(points)>0:
    clustern = getcluster_from_point(r,points[0])
    #print(clustern,points[0])
    points = [p for p in points if p not in clustern]
    #print(points, clustern, len(points))
    clusters.append(clustern)
  clusters = sorted(clusters, key=lambda x: x[0])
  return clusters
#минимизация количества команд
def get_min(m, r):
  cls = get_clusters(r)
  cum = 0
  for cl in cls:
    cum+=int(math.trunc(len(cl)/m))
    if len(cl) %m >0:
      cum+=1
  return str(cum)
#обработка входа и выдача результата
def solve(dataset):
  r,m = ast.literal_eval(dataset)
  return get_min(m, r)
```
